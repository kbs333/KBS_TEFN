```python
import pandas as pd

trading_volume_df = pd.read_csv("trading_volume.csv", index_col = "날짜", parse_dates=True)
trading_volume_df = trading_volume_df.sort_index() # 정렬 sort_values()
# trading_volume_df = trading_volume_df['2020-01-01':'2020-09-28']
trading_volume_df
```

```python
def peak_day(stock_vol, months = 6):
    df_final = pd.DataFrame()
    first_date = stock_vol.index[0]
    end_date = stock_vol.index[-1]
    date_check = -1
    
    while end_date != date_check:
        last_date = first_date + pd.DateOffset(months = months)
        
        df = stock_vol[first_date:last_date]
        date_check = df.index[-1]
        df = df.sort_values(by=['거래량'])
        df = df[:int(len(df) * 0.05)].sort_index()
        df_final = pd.concat([df_final, df])
        
        first_date = last_date + pd.DateOffset(days = 1)
    return df_final
        
peak_day_df = peak_day(trading_volume_df, months=6)
peak_day_df
```

```python
import pandas as pd

trading_volume_df = pd.read_csv("d_score.csv", parse_dates=True)
trading_volume_df = trading_volume_df.sort_index().drop('Unnamed: 0', axis=1)
trading_volume_df
```

```python
def peak_day(stock_vol, months = 6):
    stock_vol['date'] = stock_vol.date.astype('datetime64[ns]')

    df_final = pd.DataFrame()
    first_date = stock_vol.date[0]
    end_date = stock_vol.iloc[-1]['date']
    date_check = -1
    
    while end_date != date_check:
        last_date = first_date + pd.DateOffset(months = months)
        
        df = stock_vol[(stock_vol.date >= first_date) & (stock_vol.date < last_date)]
        date_check = df.iloc[-1]['date']
        df = df.sort_values(by=['d_score'])
        df = df[:int(len(df) * 0.05)].sort_values(by=['date'])
        df_final = pd.concat([df_final, df])
        
        first_date = last_date
    return df_final.reset_index(drop=True)
        
peak_day_df = peak_day(trading_volume_df, months=6)
peak_day_df
```

```python
import numpy as np
import pandas as pd
from datetime import datetime

topic_count_df = pd.DataFrame(np.random.randint(1,10,size=(2099, 9)), columns=list('123456789'))
# df.index = pd.date_range(end = "2020-10-01", periods = 100).tolist()
topic_count_df.index = pd.date_range(start = '2015-01-01', end = "2020-09-29")
topic_count_df
```

```python
def FVE(topic_count, peak_day, per=0.5):
    peak_count_df = pd.merge(peak_day, topic_count, right_index=True, left_index=True).drop(["거래량"], axis=1)
    df = pd.DataFrame({'Topic' : peak_count.columns, 'FVE': round(peak_count.sum()/peak_count.sum().sum(), 5)})
    df = df.sort_values('FVE', ascending=False).reset_index(drop=True)
    df['Cumsum'] = df.FVE.cumsum()
    df = df[df.Cumsum < per]
    return df

FVE(topic_count_df, peak_day_df)
```
